##  Non-local Means Denoise Algorithm

### Parameters
fastNlMeansDenoisingColored( src, dst, h, hcolor, templateWindowSize, searchWindowSize)
<li> src: the input image that we want to do noise reduction.
    
<li>dst: the destination if we want to export the result.
    
<li>h: the luminance component. (Larger h value removes more noise but can also remove image details).
    
<li>hcolor: the color component. (10 is the recommended value by the documentation for colored images).
    
<li>templateWindowSize: The pixel size of the area that the function will smooth. It should be an odd

<li>integer number. (7 is the recommended value by the documentation, which will fit for most cases).

<li>searchWindowSize: The pixel size of the area that the function will find and reference to. Affect 
<li>performance linearly: greater searchWindowsSize means greater denoising time. Also it should be an 

<li>odd integer number. (21 is the recommended value by the documentation, which will fit for most cases).

```python
dst1 = cv.fastNlMeansDenoisingColored(img,None,3,3,7,21) 
dst2 = cv.fastNlMeansDenoisingColored(img,None,5,5,7,21) 
dst3 = cv.fastNlMeansDenoisingColored(img,None,10,10,7,21) 
dst4 = cv.fastNlMeansDenoisingColored(img,None,3,3,5,13)
```

```python
cv.imwrite('one_1.png', dst1) 
cv.imwrite('one_2.png', dst2) 
cv.imwrite('one_3.png', dst3) 
cv.imwrite('one_4.png', dst4)
```
